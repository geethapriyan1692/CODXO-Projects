```python
## using  oversampling technique
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy=0.4)
oversampled_X, oversampled_Y = sm.fit_resample(df.drop('Class', axis=1), df['Class'])
oversampled = pd.concat([pd.DataFrame(oversampled_Y), pd.DataFrame(oversampled_X)], axis=1)
```

```python
fraud = oversampled[oversampled['Class'] == 1]
valid = oversampled[oversampled['Class'] == 0]
print('Fraud Cases: {}'.format(len(oversampled[oversampled['Class'] == 1])))
print('Valid Transactions: {}'.format(len(oversampled[oversampled['Class'] == 0])))
```
